```python
# 1. 필요한 모듈 import
import os
import random
import cv2
import numpy as np
import shutil
from matplotlib import pyplot as plt
from PIL import Image


# 2. 어종별 데이터 확인
fishes = ['blue_gill','carp','channa_argus', 'large_mouth_bass', 'crucian','catfish']


# 2-1) 어종별 file path 어종 list에 저장,
blue_gill_img = []
carp_img = []
channa_argus_img = []
large_mouth_bass_img = []
crucian_img = []
catfish_img = []

fish_imgs = [blue_gill_img, carp_img, channa_argus_img, large_mouth_bass_img, crucian_img, catfish_img]

whole_cnt = 0 # 전체 이미지 데이터 갯수

for i in range(len(fishes)):
    fish = fishes[i]
    img_list = fish_imgs[i]
    file_path = './' + fish + '_image'
    file_list = os.listdir(file_path)
    cnt = 0
    for file in file_list:
        img_list.append(file_path + '/' + file)
        cnt +=1
    whole_cnt += cnt


# 2-2) 데이터 갯수 분포 확인
for i in range(len(fishes)):
    print('-----------------------------------------------------------------')
    print('{} image count : '.format(fishes[i]), len(fish_imgs[i]))
    print('{} image count percent : '.format(fishes[i]), int((len(fish_imgs[i])/whole_cnt)*100) ,'%')
    


# 3. 데이터 정리

# Training, Test로 폴더 분류

blue_gill_train_folder = 'dataset/test_set/blue_gill'
carp_train_folder = 'dataset/test_set/carp'
channa_argus_train_folder = 'dataset/test_set/channa_argus'
large_mouth_bass_train_folder = 'dataset/test_set/large_mouth_bass'
crucian_train_folder = 'dataset/test_set/crucian'
catfish_train_folder = 'dataset/test_set/catfish'

blue_gill_test_folder = 'dataset/training_set/blue_gill'
carp_test_folder = 'dataset/training_set/carp'
channa_argus_test_folder = 'dataset/training_set/channa_argus'
large_mouth_bass_test_folder = 'dataset/training_set/large_mouth_bass'
crucian_test_folder = 'dataset/training_set/crucian'
catfish_test_folder = 'dataset/training_set/catfish'

folders = [blue_gill_train_folder, carp_train_folder, channa_argus_train_folder, large_mouth_bass_train_folder,\
    crucian_train_folder, catfish_train_folder, blue_gill_test_folder,carp_test_folder, channa_argus_test_folder, large_mouth_bass_test_folder, \
        crucian_test_folder, catfish_test_folder]

# 폴더 초기화 
for folder in folders:
    if os.path.exists(folder):
        shutil.rmtree(folder)

# 데이터 삽입
for folder in folders:
    if not os.path.exists(folder):
        os.makedirs(folder)




# 각 어종의 image data folder 에서 80%는 training folder에, 20%는 test folder에 복사하기.
for i in range(len(fishes)):

    fish_name = fishes[i]
    fish_file_list = fish_imgs[i]

    train_destination_folder = 'dataset/training_set/{}'.format(fish_name)
    test_destination_folder = 'dataset/test_set/{}'.format(fish_name)

    #랜덤으로 리스트내의 파일 섞어줌
    random.shuffle(fish_file_list) 

    # 80%는 train으로, 20%는 test로 복사함.
    for i in range(len(fish_file_list)):
        fish_file = fish_file_list[i]
        file_last_name = fish_file.split('/')[-1]
        if i < int(len(fish_file_list)*0.8): # 80% 이전은 train 폴더로 이동
            origin = fish_file
            copy = train_destination_folder + '/' + file_last_name
            shutil.copy(origin, copy)
        else: # 나머지는 test 폴더로 이동
            origin = fish_file
            copy = test_destination_folder + '/' + file_last_name
            shutil.copy(origin, copy)



```
