```python
# 1. 필요한 모듈 import
from PIL import Image
import os
import cv2

# 2. 각 어종 별 이미지 사이즈 확인, 사이즈 균일하게 자르기
# 원하는 사이즈 : 500x500
# 2-1) width와 heigth 중 더 큰 값을 500에 맞추고, 더 짧은 값은 비율에 맞게 축소한다.
# 2-2) 2-1)과 같이 진행 할 경우, 기존의 이미지가 정사각형이 아니라면, 공백이 남게 되는데 공백은 흰색으로 처리해준다.
# 2-3) 이미지 데이터가 적은 관계로 회전 시킨 이미지를 추가함.
# 2-4) 이미지 데이터가 적은 관계로 반전 시킨 이미지를 추가함.

# 어종 리스트 생성 
fishes = ['blue gill','carp','Channa argus', 'large_mouth_bass', 'crucian','catfish']

# 원하는 정사각형 이미지 사이즈
length = 500

for fish in fishes:
    # 기존 파일 불러오기
    file_path = './' + fish + '_image'
    file_list = os.listdir(file_path)
    for file in file_list:
        file_name = file_path + '/' + file

        if file_name.split('_')[-1] != 'reshaped.jpg' and file_name.split('_')[-1] != 'resized.jpg':
            img1 = cv2.imread(file_name, cv2.IMREAD_COLOR)
            image_shape = img1.shape
            width, height, color = image_shape

            # 2-1) width와 heigth 중 더 큰 값을 500에 맞추고, 더 짧은 값은 비율에 맞게 축소혹은 확대 한다.
            if width >= height:
                f_x = length/width
                f_y = f_x
            else:
                f_y = length/height
                f_x = f_y

            reshaped_img = cv2.resize(img1, dsize=(0, 0), fx=f_x, fy=f_y, interpolation=cv2.INTER_LINEAR)

            file_name_original = file_name.split('.')[1] 
            reshaped_file_name = '.' + file_name_original + '_reshaped.jpg'
            if not os.path.exists(reshaped_file_name):
                cv2.imwrite(reshaped_file_name,reshaped_img)

            # 2-2) 정사각형 이미지로 변환하고 빈 부분은 흰색으로 처리한다.

            img2 = Image.open(reshaped_file_name) # reshaped 이미지 불러오기
            img_size = img2.size #이미지의 크기 측정 
            x = img_size[0] # width
            y = img_size[1] # height

            size = max(x, y)    
            resized_img = Image.new(mode = 'RGB', size = (size, size), color = (256,256,256))
            offset = (round((abs(x - size)) / 2), round((abs(y - size)) / 2))
            resized_img.paste(img2, offset)
            resized_file_name = '.' + file_name_original + '_resized.jpg'
            if not os.path.exists(resized_file_name):
                resized_img.save(resized_file_name)

            
            # 2-3) 회전 시킨 이미지 추가
            # 정사각형 이미지 불러오기
            img3 = cv2.imread(resized_file_name, cv2.IMREAD_COLOR)

            # 회전시키기
            rot_90_img = cv2.rotate(img3, cv2.ROTATE_90_CLOCKWISE)
            rot_180_img = cv2.rotate(img3, cv2.ROTATE_180)
            rot_270_img = cv2.rotate(img3, cv2.ROTATE_90_COUNTERCLOCKWISE)

            # 저장하기
            rot_90_file_name = '.' + file_name_original + '_rot90.jpg'
            rot_180_file_name = '.' + file_name_original + '_rot180.jpg'
            rot_270_file_name = '.' + file_name_original + '_rot270.jpg'

            if not os.path.exists(rot_90_file_name):
                cv2.imwrite(rot_90_file_name,rot_90_img)

            if not os.path.exists(rot_180_file_name):
                cv2.imwrite(rot_180_file_name,rot_180_img)

            if not os.path.exists(rot_270_file_name):
                cv2.imwrite(rot_270_file_name,rot_270_img)    


            # 2-4) 반전 시킨 이미지 추가
            # 정사각형 이미지 그대로 사용
            right_left_img = cv2.flip(img3, 1)
            up_down_img = cv2.flip(img3, 0)

            # 저장하기
            right_left_file_name = '.' + file_name_original + '_rightleft.jpg'
            up_down_file_name = '.' + file_name_original + '_updown.jpg'

            if not os.path.exists(right_left_file_name):
                cv2.imwrite(right_left_file_name,right_left_img)

            if not os.path.exists(up_down_file_name):
                cv2.imwrite(up_down_file_name,up_down_img)











```
